Part B --- KNN

- Prepare the Sets

```python
X_train, y_train = data_splits["train"]
X_valid, y_valid = data_splits["valid"]
X_test,  y_test  = data_splits["test"]
```

- Sweep the K value

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

ks = [1, 3, 5, 7, 9]
valid_scores = []

for k in ks:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    acc = knn.score(X_valid, y_valid)   # Classification in Validation
    valid_scores.append(acc)
    print(f"k = {k}, Validation accuracy = {acc:.4f}")

# Best k
best_k = ks[np.argmax(valid_scores)]
print(f"\nBest k (k*): {best_k}")
```

- Plot of 'Validation Curve'

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.plot(ks, valid_scores, marker='o', linestyle='-', color='blue')
plt.title('Validation curve for KNN')
plt.xlabel('k (number of neighbours)')
plt.ylabel('Validation classification rate')
plt.grid(True)
plt.show()
```

- Assess k , k - 2 and k + 2 in the test group

```python
neighbours_to_check = [best_k - 2, best_k, best_k + 2]
neighbours_to_check = [k for k in neighbours_to_check if k > 0]  # evitar k<=0

for k in neighbours_to_check:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    test_acc = knn.score(X_test, y_test)
    print(f"k = {k}, Test accuracy = {test_acc:.4f}")
```

PART C --- CV (Cross-Validation)

- Prepare the Data and the Environment

```python
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt

X_train, y_train = data_splits["train"]
```

- Run cross-validation for each k

```python
ks = [1, 3, 5, 7, 9]
kf = KFold(n_splits=5, shuffle=True, random_state=42)

avg_train_err = []
avg_val_err = []

for k in ks:
    train_errs, val_errs = [], []
    for tr_idx, val_idx in kf.split(X_train):
        X_tr, X_val = X_train[tr_idx], X_train[val_idx]
        y_tr, y_val = y_train[tr_idx], y_train[val_idx]

        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_tr, y_tr)

        train_errs.append(1 - knn.score(X_tr, y_tr))
        val_errs.append(1 - knn.score(X_val, y_val))

    avg_train_err.append(np.mean(train_errs))
    avg_val_err.append(np.mean(val_errs))
    print(f"k={k}: mean train error={np.mean(train_errs):.4f}, mean val error={np.mean(val_errs):.4f}")
```

- View the Mean Error (Train vs Valid)

```python
x = np.arange(len(ks))
width = 0.35

plt.figure(figsize=(7,5))
plt.bar(x - width/2, avg_train_err, width, label='Train error')
plt.bar(x + width/2, avg_val_err, width, label='Validation error')
plt.xticks(x, ks)
plt.xlabel('k (neighbours)')
plt.ylabel('Average error')
plt.title('5-Fold Cross-Validation: Train vs Validation Error')
plt.legend()
plt.show()
```

PART E --- PCA

- Adjust PCA about the group of training

```python
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

X_train, y_train = data_splits["train"]
X_valid, y_valid = data_splits["valid"]
X_test,  y_test  = data_splits["test"]

pca = PCA().fit(X_train)  # ajusta PCA con train
explained = pca.explained_variance_ratio_
cum = np.cumsum(explained)

plt.figure(figsize=(7,4))
plt.plot(np.arange(1, len(cum)+1), cum, marker='o')
plt.axhline(0.95, linestyle='--', color='red', label='95% variance')
plt.xlabel('Number of principal components')
plt.ylabel('Cumulative explained variance')
plt.title('Scree Plot (PCA)')
plt.legend()
plt.grid(True)
plt.show()
```

- Transform Train-Valid-Test into a M components

```python
m = np.argmax(cum >= 0.95) + 1   # número de componentes que alcanza 95 %
print("Número de componentes seleccionados:", m)

pca_m = PCA(n_components=m).fit(X_train)
Z_train = pca_m.transform(X_train)
Z_valid = pca_m.transform(X_valid)
Z_test  = pca_m.transform(X_test)
```

Training KNN (KNN with PCA)

```python
ks = [1, 3, 5, 7, 9]
valid_scores_pca = []

for k in ks:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(Z_train, y_train)
    acc = knn.score(Z_valid, y_valid)
    valid_scores_pca.append(acc)
    print(f"k={k}, Validation accuracy (PCA)={acc:.4f}")

best_k_pca = ks[np.argmax(valid_scores_pca)]
print(f"\nBest k after PCA: {best_k_pca}")
```

```python
plt.plot(ks, valid_scores_pca, marker='o')
plt.title('Validation Curve after PCA')
plt.xlabel('k (neighbours)')
plt.ylabel('Validation Accuracy')
plt.grid(True)
plt.show()
```

- Assess in test (KNN + PCA)

```python
knn_pca = KNeighborsClassifier(n_neighbors=best_k_pca)
knn_pca.fit(Z_train, y_train)
test_acc_knn_pca = knn_pca.score(Z_test, y_test)
print(f"Test accuracy (KNN + PCA) = {test_acc_knn_pca:.4f}")
```

- Naïve Bayes with PCA

```python
nb_pca = GaussianNB()
nb_pca.fit(Z_train, y_train)

train_acc_nb_pca = nb_pca.score(Z_train, y_train)
valid_acc_nb_pca = nb_pca.score(Z_valid, y_valid)
test_acc_nb_pca  = nb_pca.score(Z_test,  y_test)

print(f"Train: {train_acc_nb_pca:.4f}, Valid: {valid_acc_nb_pca:.4f}, Test: {test_acc_nb_pca:.4f}")
```
